## Line Style

````python
x = [2, 3, 5, 6]
y = np.array([4, 1, 10, 2])

#linestyle -(solid), -- dashed, -. dashdotted, 

plt.plot(x, y, linestyle = 'solid', linewidth = 5)
plt.plot(x, y * 0.5, linestyle = 'dashed', linewidth = 5)
plt.plot(x, y * 0.25, linestyle = 'dotted', linewidth = 5)
plt.plot(x, y * 0.25, linestyle = 'dashdot', linewidth = 5)
plt.plot(x, y * 0.75, linestyle = ':', linewidth = 5)
plt.show()
````

````python
x = np.arange(-25, 25)
y = 2 * x
plt.plot(x, y, c='g', linewidth = 2) # c ='g' or c ='green' 
plt.plot(x, -2 * x, c='b', linestyle='dashed', linewidth = 2) # c ='b' or c ='blue' 
plt.plot(np.zeros(100),  np.arange(-50, 50), c='r', linestyle='dotted', linewidth = 2) # c ='r' or c ='red' 
plt.plot(x,np.zeros(50), c='y', linestyle='dashdot', linewidth = 2) # c ='y' or c ='yellow' 
plt.title('X versus Y')
plt.show()
````

````python
x = np.arange(-25, 25)
plt.plot(x, 2 * x, c='r', linewidth = 2) # c ='r' or c ='red' 
plt.plot(x, -2 * x, c='g', linestyle='dotted', linewidth=2) # c ='g' or c ='green' 
plt.plot(x, x ** 2, c='b', linestyle='dashed', linewidth=2) # c ='b' or c ='blue'
plt.plot(x, -(x ** 2), c='y', linestyle='dashdot', linewidth=2) # c ='y' or c ='yellow' 
plt.title('X versus Y')
plt.show()
````

## Bar Graph

### Creating a simple Bar graph

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data)
plt.show()
````

## Adding color to Bar Graph

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color = 'coral')
plt.show()
````

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color = 'coral', alpha = 0.75)
plt.show()
````

## Adding multiple colors

````python
colors = ['red','blue','yellow','green','coral']
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color = colors, alpha = 0.75)
plt.show()
````

### Adding name to x-axis value

````python
colors = ['red','blue','yellow','green','coral']
data = [20, 50, 60, 80, 10]
x = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, data, color = colors, alpha = 0.75)
plt.show()
````

### Adding Label to Bar graph

````python
colors = ['red','blue','yellow','green','coral']
data = [20, 50, 60, 80, 10]
x = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, data, color = colors, alpha = 0.75)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('X versus Y')
plt.show()
````

### Adding Grid to Bar Graph

````python
colors = ['red','blue','yellow','green','coral']
data = [20, 50, 60, 80, 10]
x = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, data, color = colors, alpha = 0.75)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('X versus Y')
plt.grid(color='green', linestyle='dotted', alpha = 0.75)
plt.show()
````

## Multiple Bar graphs

````python
data1 = np.random.randint(0, 100, size=5)
data2 = np.random.randint(0, 100, size=5)
data3 = np.random.randint(0, 100, size=5)
x = np.array([1, 2, 3, 4, 5])
width = 0.25

plt.bar(x, data1, width = width,  color = '#7be382')
plt.bar(x + width, data2, width = width, color = '#eedc82')
plt.bar(x + width * 2, data3, width = width, color = '#cd5c5c')
plt.show()
````

### Stacked Bar graphs

````python
data1 = np.random.randint(0, 100, size=5)
data2 = np.random.randint(0, 100, size=5)
data3 = np.random.randint(0, 100, size=5)
x = np.array([1, 2, 3, 4, 5])
width = 0.25

plt.bar(x, data1)
plt.bar(x , data2)
plt.bar(x , data3)
plt.show()
````

````python
np.random.seed(1)
data1 = [75,74, 98, 78, 41]
data2 = [10,20, 25, 40, 25]
data3 = [20, 10, 30, 30, 35]
width = 0.35       # the width of the bars: can also be len(x) sequence

print(data1, data2, data3)

fig, ax = plt.subplots()
ax.bar(x, data1, label='Men')
ax.bar(x, data2, 
       label='Women', bottom=data1)
ax.bar(x, data3,
       label='Other', bottom=data1)

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
````

````python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

     # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means,  label='Men')
ax.bar(labels, women_means, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
````

````python

# np.random.normal(mu, sigma, size)
# mu -> mean, 
# sigma -> standard deviation
# size - number of obsevations or number of data points

normal_array = np.random.normal(28, 10, 100000)
plt.hist(normal_array,bins=50, color='lightgreen', density=True)
plt.show()
````
