```python
tastes = ['Fruity', 'Earthy', 'Spicy', 'Smoky','Flowery', 'Acidic']
taste_count = []

for taste in tastes:
    num = len(df[df['description'].str.contains(taste.lower())])
    taste_count.append(num)
    
print(taste_count)

pd.Series(taste_count, index = tastes).plot(kind='bar')
plt.show()


        
```
