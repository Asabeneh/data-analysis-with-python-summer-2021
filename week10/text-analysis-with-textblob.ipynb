```python
## polarity measure: -1 to 1


lst = [
    'I love people',
    'i like everyone', 
    'I enjoy using Python and JavaScript', 
    'I do not really like what I see every day', 
    'It is has been going very bad on daily basis.',
    'It has been nice everyday',
    'It is raining'
]
```

```python
df = pd.DataFrame(lst, columns=['sentence'])
```

```python
df
```

```python
for sent in lst:
    blob = TextBlob(sent)
    sentiment = blob.sentiment
    print(sentiment, sentiment.polarity)
```

```python
def find_polarity (txt):
    blob = TextBlob(str(txt))
    sentiment = blob.sentiment
    return sentiment.polarity
```

```python
df['polarity'] = df['sentence'].apply(find_polarity)
```

```python
df
```

```python
def find_sentiment(txt):
    if find_polarity(txt) > 0:
        return 'Postive'
    # -0.25  to 0.25
    elif find_polarity(txt) == 0:
        return 'Neutral'
    elif find_polarity(txt) < 0:
        return 'Negative'
    else:
        return 'Unknow'
```

```python
df['sentiment'] = df['sentence'].apply(find_sentiment)
```

```python
df
```

```python
df['sentiment'].value_counts().plot(kind='bar')
plt.show()
```

```python
def draw_bar_graph (df, col, title='Sentiment'):
    df[col].value_counts().plot(kind='bar', color = ['#2fd08e','#edc712','#ee3f11'], title=title)
    plt.show()
```

```python
draw_bar_graph(df, 'sentiment')
```

```python
# Lets make use of the most powerful natural language processing package, nltk
```

```python
sentences = 'I love people. People are great. Are not they ? If you do not love people what else can you love. Love is the most important thing.'
```

```python
sent_tokens = nltk.tokenize.sent_tokenize(sentences)
```

```python
sent_tokens
```

```python
word_tokens = nltk.tokenize.word_tokenize(sentences)
```

```python
word_tokens
```

```python
stop_words = nltk.corpus.stopwords.words('english')
print(stop_words)
```

```python
words = [w for w in word_tokens if w.isalpha()]
words
```

```python
words_without_stopwords = [w.lower() for w in word_tokens if w.isalpha() and w.lower() not in stop_words]
```

```python
words_without_stopwords
```

```python
freq_dist = nltk.probability.FreqDist(words_without_stopwords)
freq_dist
```

```python
freq_dist.most_common(3)
```

```python
freq_dist.plot()
plt.show()
```

```python
tokens = [w.lower() for w in nltk.tokenize.word_tokenize(texts) if w.isalpha() and w.lower() not in stop_words]
```

```python
freq_dist = nltk.probability.FreqDist(tokens)
```

```python
freq_dist
```

```python
freq_dist.most_common()
```

```python
freq_dist.plot(25)
plt.show()
```
